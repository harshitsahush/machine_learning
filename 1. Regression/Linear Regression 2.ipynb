A linear regression model from [Data Professor](https://www.youtube.com/watch?v=R15LjD8aCzc).
Data from [Data Professor github](https://github.com/dataprofessor).


```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/BostonHousing.csv")
print(df)
```

```python
y = df["medv"]
X = df.drop("medv", axis = 1)
print(y)
print(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)

print(lr_train_mse)
print(lr_train_r2)
print(lr_test_mse)
print(lr_test_r2)
```
