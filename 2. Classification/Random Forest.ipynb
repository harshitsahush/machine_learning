```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/harshitsahush/machine_learning_datasets/main/500hits.csv", encoding = 'latin-1')
df = df.drop(['PLAYER', 'CS'], axis=1)    #remove unnecessary cols

df.head()
```

```python
y = df['HOF']
X = df.drop('HOF', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 69)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(max_depth = 4, criterion = 'entropy')
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy score :", accuracy)
```
